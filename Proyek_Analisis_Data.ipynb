## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana pengaruh hari libur terhadap minat user untuk bersepeda?


```python
import seaborn as sns

sns.barplot(y=hour_df["cnt"], x=hour_df["holiday"], orient="v", color='#cc307a')
plt.ylabel("Count of Users")
plt.xlabel("Holiday(0 : Bukan Hari Libur Nasional, 1 : Hari Libur Nasional)")
plt.title("The influence of holidays on users' interest in cycling")
plt.show()
```

### Pertanyaan 2: Bagaimana pengaruh musim tahunan terhadap durasi user dalam bersepeda tiap jamnya?

```python
avg = hour_df.groupby(['season'])['hr'].mean().reset_index()
seasons_label = {1:'Spring', 2:'Summer', 3:'Fall', 4:'Winter'}
plt.figure(figsize=(7, 7))
plt.pie(avg.hr, labels=avg.season.map(seasons_label), autopct='%1.1f%%', startangle=90, colors=['#f7e0b8','#bbeed7','#b2e6e9','#f7cde1'])
plt.title('Average Bike Rentals by Season')
plt.axis('equal')
plt.show()
```

**Insight:**
- Hasil visualisasi data pertama menunjukkan bahwa tingkat penyewaan sepeda pada hari libur nasional lebih rendah daripada hari biasa.
- Hasil visualisasi data kedua menunjukkan bahwa durasi user bersepeda tiap jamnya pada musim spring menurun daripada musim lainnya.

## Analisis Lanjutan (Opsional)

```python
avg_windspeed = hour_df.groupby('windspeed')['cnt'].mean()

plt.figure(figsize=(10,6))
plt.plot(avg_windspeed.index, avg_windspeed.values, marker='o', linestyle='-', color='#308acc')

plt.title('Effect of Windspeed on Bike Rentals', fontsize=16)
plt.xlabel('Windspeed', fontsize=14)
plt.ylabel('Bike Rentals (cnt)', fontsize=14)
plt.grid(True)
plt.legend()

plt.show()
```

Insight :
- Grafik memperlihatkan minat penyewa sepeda cenderung menurun ketika kecepatan angin sedang tinggi. Penyewa terlihat lebih nyaman bersepeda ketika windspeed berada di antara 0.2 dan 0.4 dimana itu berarti kecepatan angin tidak terlalu rendah ataupun tinggi.
